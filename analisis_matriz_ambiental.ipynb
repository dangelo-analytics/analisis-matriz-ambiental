# **DESCRIPCION DEL DATASET ORIGINAL**

El dataset corresponde a una matriz de seguimiento de cumplimiento ambiental, cuyo objetivo es consolidar en un único archivo los distintos compromisos y requisitos legales asociados a una operación.

La estructura original presenta un enfoque monolítico, donde múltiples tipos de información (permisos, estados, responsables, fechas y observaciones) se agrupan en una sola tabla con atributos ambivalentes que dificultan su análisis.

Adicionalmente, existen hojas complementarias que replican información a mayor nivel de detalle por componente ambiental, generando redundancia y posibles inconsistencias.

## **OBJETIVO DEL DATA ANALISIS REALIZADO**



1.   Transformar estructura caotica a modelo analitico
2.   Reducir redundancia
3.   Facilitar el analisis del cumplimiento
4.   Generar KPIs claros
5.   Creacion  de Dashboard en Power BI

## **TRANSFORMACIONES**

“Despues de la exploracion inicial se procede con la limpieza, normalización y estructuración del dataset para su posterior análisis.”

Se identificaron múltiples deficiencias en el diseño original de la tabla, incluyendo:

- Violaciones a la 1NF (atributos no atómicos)
- Redundancia semántica entre columnas
- Uso de descripciones en lugar de categorías
- Columnas sin valor analítico

Se procede a rediseñar la estructura del dataset, eliminando columnas irrelevantes, renombrando variables para mayor claridad y creando nuevas columnas orientadas a un modelo relacional futuro (Documentos y Observaciones).

### **LIMPIEZA DE FILAS Y COLUMNAS VACIAS O SUCIAS**

Se eliminaran las ultimas 2 filas de la tabla, y las 2 ultimas columnas hacia la derecha
 ya que no poseen valores, o los resulatdos totales bajo las columnas son irrelevantes
 e incompletos

```python

df = df.iloc[:-2]
df = df.iloc[:,:-2]
```

```python
df.head()
```

```python
df.tail(5)
```

###**ELIMINACION DE COLUMNAS SEGUN NUEVO MODELO**

Periodicidad no es relevante para el analisis de cumplimiento. Sólo se requiere conteo de entregas respecto al total. Tipo de tarea esta mal asociado conceptualmente durante la concepcion del modelo o quedó mal etiquetado.

```python
df = df.drop(columns=['Periodicidad', 'Tipo de Tarea'])
```

```python
df
```

### **RENOMBRAR COLUMNAS**

```python
df = df.rename(columns={
    'N°': 'ID',
    'Resultado':'Entregados',
    'Faena':'Actividad',
    'Descripcion ':'Requisito',
    'Documento /Accion':'Detalle_doc'
})
```

```python
df
```

### **CREAR NUEVAS COLUMMNAS**

```python
df['Documento'] = None
df['Doc_ID'] = None
df['Obs_ID'] = None
```

### TRANSFORMAR COLUMNA OBSERVACIONES A BOLEANO

```python
df['Observaciones'] = df['Observaciones'].notna()
```

### **LIMPIEZA DE ENCABEZADO**

```python
df.columns = df.columns.str.strip()
```

### **REORDENAR COLUMNAS**

```python
df = df[
    ['ID',
     'Etapa Proyecto',
     'Actividad',
     'Area Responsable',
     'Componente',
     'Requisito',
     'Documento',
     'Detalle_doc',
     'Doc_ID',
     'Entregables',
     'Entregados',
     'Observaciones',
     'Obs_ID']
]
```
